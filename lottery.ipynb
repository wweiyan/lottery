```python
#https://www.17500.cn/getData/ssq.TXT
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
datas = pd.read_table('ssq.TXT',
                      sep='\s+',
                      header=None,
                      names=['id','date',
                             'b1','b2','b3','b4','b5','b6',
                             'r1',
                             'o1','o2','o3','o4','o5','o6','o7','o8','o9','o10','o11','o12','o13','o14','o15','o16','o17','o18','o19','o20'])
lottery_datas = datas[['id', 'date', 'b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'r1']]
```

```python
m = len(lottery_datas)
b1_mode = []
for i in range(m):
    b1_mode.append(lottery_datas['b1'][0:i+1].mode()[0])
plt.plot(range(m),b1_mode)
```

```python
import matplotlib.font_manager
font_set = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\simsun.ttc')
m = len(lottery_datas)

b = ['b1', 'b2', 'b3', 'b4', 'b5', 'b6']
plt.figure(figsize=(16,8))
for i,j in zip(range(6),b):
    plt.subplot(2,3,i+1)
    mode = []
    for k in range(m):
        mode.append(lottery_datas[j][0:k+1].mode()[0])
    plt.plot(range(m),mode)
    plt.title("众数是%d"%lottery_datas[j].mode()[0],fontproperties=font_set)
plt.show()
```
